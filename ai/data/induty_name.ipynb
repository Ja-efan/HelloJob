```python
industries_df = pd.read_excel("한국표준산업분류_수정본.xlsx")
industries_df.head()
```

```python
# null이 아닌 값에 대해 패딩을 추가하는 함수 정의
def add_padding(value, target_length, pad_char='0'):
    if pd.isna(value):
        return value
    # 소수점 제거하고 정수 부분만 사용
    int_value = str(int(float(value)))
    # 왼쪽에 패딩 추가
    return pad_char * (target_length - len(int_value)) + int_value

# 각 분류 코드 컬럼에 패딩 적용
industries_df['중분류 코드'] = industries_df['중분류 코드'].apply(lambda x: add_padding(x, 2) if pd.notna(x) else x)
industries_df['소분류 코드'] = industries_df['소분류 코드'].apply(lambda x: add_padding(x, 3) if pd.notna(x) else x)
industries_df['세분류 코드'] = industries_df['세분류 코드'].apply(lambda x: add_padding(x, 4) if pd.notna(x) else x)
industries_df['세세분류 코드'] = industries_df['세세분류 코드'].apply(lambda x: add_padding(x, 5) if pd.notna(x) else x)
```

```python
def map_industry_code_to_name(stocked_list_with_induty_code, industries_df):
    # 결과를 저장할 리스트 초기화
    industry_names = []
    
    # 각 행에 대해 처리
    for induty_code in stocked_list_with_induty_code['induty_code']:
        # 코드가 없는 경우 처리
        if pd.isna(induty_code):
            industry_names.append('기타')
            continue
        
        # 코드를 문자열로 변환
        if not isinstance(induty_code, str):
            # 소수점 제거 (예: 1.0 -> '1')
            if isinstance(induty_code, float):
                induty_code = str(int(induty_code))
            else:
                induty_code = str(induty_code)
        
        code_length = len(induty_code)
        
        # 코드 길이에 따라 분기 처리
        if code_length == 2:
            # 중분류 코드에서 찾기
            match = industries_df[industries_df['중분류 코드'].astype(str) == induty_code]
            if not match.empty:
                industry_names.append(match.iloc[0]['중분류 항목명'])
            else:
                industry_names.append('기타')
                
        elif code_length == 3:
            # 소분류 코드에서 찾기
            match = industries_df[industries_df['소분류 코드'].astype(str) == induty_code]
            if not match.empty:
                industry_names.append(match.iloc[0]['소분류 항목명'])
            else:
                industry_names.append('기타')
                
        elif code_length == 4:
            # 세분류 코드에서 찾기
            match = industries_df[industries_df['세분류 코드'].astype(str) == induty_code]
            if not match.empty:
                industry_names.append(match.iloc[0]['세분류 항목명'])
            else:
                industry_names.append('기타')
                
        elif code_length == 5:
            # 세세분류 코드에서 찾기
            match = industries_df[industries_df['세세분류 코드'].astype(str) == induty_code]
            if not match.empty:
                industry_names.append(match.iloc[0]['세세분류 항목명'])
            else:
                industry_names.append('기타')
                
        else:
            # 코드 길이가 2~5 범위를 벗어나면 '기타'로 설정
            industry_names.append('기타')
    
    # 결과 리스트를 데이터프레임에 새 컬럼으로 추가
    stocked_list_with_induty_code['induty_name'] = industry_names
    
    return stocked_list_with_induty_code

# 함수 사용
stocked_list_with_induty_code = map_industry_code_to_name(sotcked_list_with_induty_code, industries_df)
```
