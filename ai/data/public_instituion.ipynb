````python
# OpenAI API를 사용하여 각 기관의 사업 종목을 반환하는 함수 정의
import openai
import time
import os
from tqdm import tqdm
from pydantic import BaseModel

def get_business_categories(df, api_key, batch_size=10):
    """
    데이터프레임의 기관 정보를 batch_size만큼 잘라서 OpenAI API에 요청하여 사업 종목을 반환하는 함수
    
    Args:
        df (DataFrame): 기관 정보가 담긴 데이터프레임
        api_key (str): OpenAI API 키
        batch_size (int): 한 번에 처리할 기관의 수
        
    Returns:
        DataFrame: 사업 종목 정보가 추가된 데이터프레임
    """
    
    system_prompt = """당신은 공공기관의 정보를 분석하여 해당 공기업의 '사업 종목'을 간결하게 추출하는 전문가입니다.
몇가지 예시를 제공하겠습니다. 예시를 참고하여 나머지 공기업들도 '사업 종목'을 추출해주세요.


# 예시

한국가스안전공사가스: 안전 관리 및 검사

한국교통안전공단: 교통안전 관리 및 자동차 검사

한국산업안전보건공단: 산업재해 예방 및 산업안전보건 사업

한국지역난방공사: 지역 냉난방 공급

한국토지주택공사: 토지 개발, 주택 건설 및 공급
```"""

    class BusinessCategory(BaseModel):
        institution_name: str  
        business_category: str

    # API 키 설정
    openai.api_key = api_key
    
    # 결과를 저장할 리스트
    business_categories = []
    
    client = openai.OpenAI(api_key=api_key)
    
    # 배치 단위로 처리
    for i in range(0, len(df), batch_size):
        batch = df.iloc[i:i+batch_size]
        batch_results = []
        
        print(f"Processing batch {i//batch_size + 1}/{(len(df) + batch_size - 1)//batch_size}")
        
        
        
        
        for _, row in tqdm(batch.iterrows(), total=len(batch)):
            # 각 기관의 정보 구성
            institution_info = f"""
            기관명: {row['기관명']}
            기관유형: {row['기관유형']}
            주무부처: {row['주무부처']}
            설립목적: {row['설립목적']}
            주요기능 및 역할: {row['주요기능 및 역할']}
            경영목표 및 전략: {row['경영목표 및 전략']}
            """
            
            try:
                # OpenAI API 요청
                response = client.beta.chat.completions.parse(
                    model="gpt-4.1-mini",
                    messages=[
                        {"role": "system", "content": system_prompt},
                        {"role": "user", "content": f"다음 공공기관의 정보를 바탕으로 주요 사업 종목을 간결하게 추출하세요.\n\n{institution_info}"}
                    ],
                    temperature=0.5,
                    max_tokens=150,
                    response_format=BusinessCategory
                )
                
                # 응답에서 사업 종목 추출
                business_category = response.choices[0].message.parsed
                batch_results.append(business_category)
                
                # API 요청 제한 방지를 위한 딜레이
                time.sleep(0.5)
                
            except Exception as e:
                print(f"Error processing {row['기관명']}: {e}")
                batch_results.append("Error: 사업 종목 정보 추출 실패")
        
        # 배치 결과 저장
        business_categories.extend(batch_results)
        
        # 배치 간 딜레이
        time.sleep(1)
    
    return business_categories
````

````python
# 함수 사용 예시 (실행시에는 API 키 입력 필요)
api_key = os.getenv("OPENAI_API_KEY")
business_categories = get_business_categories(df, api_key)
````
